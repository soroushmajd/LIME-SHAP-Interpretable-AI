```python
from PIL import Image
import tensorflow as tf
import tensorflow
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2, preprocess_input
import numpy as np
import keras.utils as image
import os
from tensorflow.keras.applications.mobilenet import preprocess_input


# Load the MobileNet V2 model pre-trained on ImageNet
model = MobileNetV2(weights='imagenet')



def transform_img_fn(paths):
    out = []
    for imgs in paths:
        img = image.load_img(imgs, target_size=(224, 224))
        i = image.img_to_array(img)
        i = np.expand_dims(i, axis=0)
        i = preprocess_input(i)
        out.append(i)
    return np.vstack(out)

input_tensor = transform_img_fn([os.path.join('data','/content/cat-dog.jpg')])


predictions = model.predict(input_tensor)
# Load the ImageNet class labels
labels_url = 'https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt'
labels_path, _ = urllib.request.urlretrieve(labels_url, 'imagenet_classes.txt')

with open(labels_path) as f:
    class_labels = f.readlines()
class_labels = [label.strip() for label in class_labels]

top_classes = np.argsort(predictions[0])[::-1][:10]
top_probabilities = predictions[0][top_classes]

for i in range(10):
    class_index = top_classes[i]
    class_label = class_labels[class_index]
    probability = top_probabilities[i]
    print(f"{class_label}: {probability:.4f}")
```

```python
from lime import lime_image
explainer = lime_image.LimeImageExplainer()
```

```python
explanation = explainer.explain_instance(input_tensor[0].astype('double'), model.predict, top_labels=7, hide_color=0, num_samples=5000)
```

```python
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt
```

```python


temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=False, num_features=10, hide_rest=False)
plt.imshow(mark_boundaries(temp / 2 + 0.5, mask))
```

```python


temp, mask = explanation.get_image_and_mask(explanation.top_labels[0], positive_only=True, num_features=5 ,hide_rest=False)
plt.imshow(mark_boundaries(temp / 2 + 0.5, mask))


```

```python

dict_heatmap = dict(explanation.local_exp[explanation.top_labels[0]])
heatmap = np.vectorize(dict_heatmap.get)(explanation.segments) 
plt.imshow(heatmap, cmap = 'RdBu', vmin  = -heatmap.max(), vmax = heatmap.max())
plt.colorbar()
plt.show()
```
